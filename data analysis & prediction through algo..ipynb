```python
diabetes_df=pd.DataFrame(diabetes.data)
diabetes_df.columns=diabetes.feature_names
diabetes_df.head()
```

```python
cancer_df=pd.DataFrame(cancer.data)
cancer_df.columns=cancer.feature_names
cancer_df.head()
```

```python
diabetes_df['diabetic_level']=diabetes.target
diabetes_df.head()
```

```python
cancer_df['cancer_level']=cancer.target
cancer_df.head()
```

```python
def diabetes_check(x):
    if x>127:
        return 1
    else:
        return 0
```

```python
diabetes_df['target']=diabetes_df['diabetic_level'].apply(diabetes_check)
```

```python
diabetes_df.head()
```

```python
x=diabetes_df.drop(['age','sex','s1','s2','s3','s4','s5','s6','diabetic_level','target'],axis=1)
x.head()
```

```python
x=diabetes_df.drop([],axis=1)
x.head()
```

```python
y=diabetes_df.target
y.head()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
from sklearn import metrics
from sklearn.neighbors import KNeighborsClassifier
```

```python
list_accuracy=[]
for i in range(1,50):
    knn=KNeighborsClassifier(n_neighbors=i)
    knn.fit(x_train,y_train)
    ypre=knn.predict(x_test)
    list_accuracy.append(metrics.accuracy_score(y_test,ypre))
```

```python
plt.plot(range(1,50),list_accuracy)
plt.grid()
```

```python
knn=KNeighborsClassifier(n_neighbors=20)
knn.fit(x_train,y_train)
ypre=knn.predict(x_test)
print(metrics.accuracy_score(y_test,ypre))
```
